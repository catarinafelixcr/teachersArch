```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Caminho dos ficheiros
base_path = "C:\\Users\\Miguel Ant√≥nio\\Desktop\\pecd2025-pl2g4\\Code\\gitlabAPI"  # Substituir pelo caminho real

# Fun√ß√£o de EDA b√°sica
def run_eda(df, interval_number):
    print(f"\n===== Intervalo {interval_number} =====")
    print("üìå Primeiras linhas:")
    print(df.head())

    print("\nüìä Informa√ß√£o geral:")
    print(df.info())

    print("\nüìà Estat√≠sticas descritivas:")
    print(df.describe())

    print("\nüîç Valores nulos:")
    print(df.isnull().sum())

    # Correla√ß√£o
    plt.figure(figsize=(10, 6))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
    plt.title(f'Mapa de Correla√ß√£o - Intervalo {interval_number}')
    plt.show()

# L√™ e executa EDA para os 5 datasets
dfs = []
for i in range(1, 6):
    path = os.path.join(base_path, f'student_features_interval_{i}.csv')
    df = pd.read_csv(path)
    dfs.append(df)
    run_eda(df, i)
```
